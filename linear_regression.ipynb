```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Boston_Housing.csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
```

```python
X=df.iloc[:,:-2]
y=df.iloc[:,-2]

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
```

```python
sas=pd.read_csv('sas_output.csv')
sas.head()

py=pd.read_csv('py_output.csv')
py.head()

r=pd.read_csv('r_output.csv')
r.head()
```

```python
fig, axs = plt.subplots(1, 3, sharey=True)
sas.plot(kind='scatter', x='test', y='test', ax=axs[0], figsize=(16, 8))
py.plot(kind='scatter', x='test', y='test', ax=axs[1])
r.plot(kind='scatter', x='test', y='test', ax=axs[2])
plt.plot(sas['predicted'],sas['predicted'],color='red')
plt.plot(py['predicted'],py['predicted'],color='green')

plt.plot(r['predicted'],r['predicted'])
```

```python
# fig, axs = plt.subplots(1, 3, sharey=True)
sas.plot(kind='scatter', x='test', y='predicted')
plt.plot(sas['predicted'],sas['predicted'],color='red')
plt.title('actual v/s predicted results using SAS')
plt.legend(loc='upper left')

py.plot(kind='scatter', x='test', y='predicted')
plt.plot(py['predicted'],py['predicted'],color='green')
plt.title('actual v/s predicted results using PYTHON')
plt.legend(loc='upper left')


r.plot(kind='scatter', x='test', y='predicted')
plt.plot(r['predicted'],r['predicted'],color='orange')
plt.title('actual v/s predicted results using R')
plt.legend(loc='upper left')
```
